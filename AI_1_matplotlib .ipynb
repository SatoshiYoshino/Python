### 　複数のグラフの描画

#### 複数のグラフを描画する場合は、２つの方法がある。
・１つの図内に全部のグラフを書く。この場合、plot関数を複数回呼び出す<br>
・各グラフを別々の図として書く。subplot関数を使う

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0,2 * np.pi,0.1)
sin = np.sin(a)
plt.subplot(2,1,1)
plt.plot(a,sin)

cos = np.cos(a)
plt.subplot(2,1,2)
plt.plot(a,cos)

#subplot（行数、列数、プロット番号）　今回の場合は、上下に分割するので行数は２、列数は１となる。　（４分割の場合は、2,2となる。）
#プロット番号で、位置の指定をする。今回は、上が１、下に表示される図が２となる。（４等文の場合、左から１、２そして左下に移動して３、４と順番にプロットされる。）
```
