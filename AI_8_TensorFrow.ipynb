```python
import tensorflow as tf
```

```python
a = tf.constant(3.0,tf.float32,name='a')
b = tf.constant(2.0,tf.float32,name='b')

x = tf.placeholder(tf.float32,name='x')
y = a*x+b

sess = tf.Session()
writer = tf.summary.FileWriter('logs', graph=sess.graph)
print(sess.run(y,{x:[0,1,2,3]}))
```

「chapter09」ディレクトリ内に上記の.pyファイルを作成して実行すると、chaputer09内にlogsが作成される。
「chapter09」ディレクトリのterminalを選択し、以下のコマンドを入力する。

cd Chapter09 <br>
