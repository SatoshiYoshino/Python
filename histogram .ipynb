```python
# グラフ化に必要なものの準備
import matplotlib
import matplotlib.pyplot as plt

# データの扱いに必要なライブラリ
import pandas as pd
import numpy as np
import datetime as dt

%matplotlib inline
```

```python
#下記のおまじない設定をしておくとチャートが綺麗にかけます
plt.style.use('ggplot') 
font = {'family' : 'meiryo'}
matplotlib.rc('font', **font)
```

```python
data=pd.read_csv('mau_sample1.1.csv')
```

```python
data.head(5)
```

```python
data.plot.bar(alpha=0.6, figsize=(12,10))
plt.title('gragh', size=16)
```

```python
data.plot.hist(figsize=(15,20),subplots=True)
#subplots=Trueでグラフを分割
#figsize=(横,縦)でグラフのサイズを指定
plt.title(u'hutuunogurahu', size=16)
```

```python
data.plot.hist(figsize=(12,20),subplots=True,normed=True)
plt.rcParams["font.size"] = 15
#plt.rcParams["font.size"] = 18で全てのフォントサイズを指定
#normed=Trueで正規化
#subplots=Trueでグラフを分割
#figsize=(横,縦)でグラフのサイズを指定
```

```python

data.plot.hist(figsize=(12,20),subplots=True,normed=True)

plt.tight_layout()
#plt.tight_layout()は、グラフの位置やサイズが自動で調整されて、出力画像からのはみ出しを抑える
plt.title("Title", fontsize=18)
plt.xlabel("xlabel", fontsize=18)
#ラベルの指定とフォントサイズを変更
plt.ylabel("ylabel", fontsize=18)
#ラベルの指定とフォントサイズを変更
plt.legend(loc='center',fontsize=18)
#plt.legendは、凡例を表示
plt.tick_params(labelsize=18)
#plt.tick_params(labelsize = 数字 )で目盛りの文字サイズを指定(tickは、目盛りを意味している)
#plt.rcParams["font.size"] = 18とすると、全ての文字サイズが変更される。
```

```python
data.plot.hist(figsize=(12,20),subplots=True,normed=True,bins=50)
plt.tight_layout()
#plt.tight_layout()は、グラフの位置やサイズが自動で調整されて、出力画像からのはみ出しを抑える
plt.title("Title", fontsize=18)
plt.xlabel("xlabel", fontsize=18)
#ラベルの指定とフォントサイズを変更
plt.ylabel("ylabel", fontsize=18)
#ラベルの指定とフォントサイズを変更
plt.legend(loc='center',fontsize=18)
#plt.legendは、凡例を表示
plt.tick_params(labelsize=18)
#plt.tick_params(labelsize = 数字 )で目盛りの文字サイズを指定(tickは、目盛りを意味している)
#plt.rcParams["font.size"] = 18とすると、全ての文字サイズが変更される。
data.plot(kind='kde', figsize=(15,5))
#正規分布作成。
```
